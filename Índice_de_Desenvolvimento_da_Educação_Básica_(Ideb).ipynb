Isso pois as correlações entre a variável `ideb` e as demais variáveis mencionadas são positivas. 

O coeficiente de correlação de Pearson avalia o grau e a direção da correlação linear entre duas variáveis. Deste modo, quanto mais próximos de 1 ou -1, maior a intensidade da correlação, o sinal indica apenas a direção (positivamente correlacionados ou negativamente correlacionados). Deve-se mencionar também que quanto mais próximo de 0 menor a força da correlação.

## Scatterplots

```python
# Create the function to visualize to bar chart
def draw_scatterplot(list_feature,df):
    fig, ax = plt.subplots(int(len(list_feature)/2),2, figsize=(15,15))
    colunas = list_feature
    indice = 0
    for j in range(len(ax)):
        for i in range(len(ax[j])):
            ax[j][i].tick_params(labelsize=12)
            ax[j][i] = sns.scatterplot(data=df, x=colunas[indice],y='ideb',ax=ax[j][i])
            indice += 1
    fig.suptitle('Gráficos de Dispersão - IDEB em função de outras variáveis', position=(.5,1), fontsize=22)
    fig.tight_layout(pad=5)
    return
list_features_regiao = num_variables_regiao.columns.drop(['ideb','ano'])
list_features_brasil = num_variables_brasil.columns.drop(['ideb','ano'])
```

```python
# df_regiao 
draw_scatterplot(list_features_regiao,df_regiao)
```

Percebemos claramente a tendência de aumento no IDEB à medida que são aumentadas as variáveis: `taxa_aprovacao`, `indicador_rendimento`, `nota_saeb_media_padronizada` e `projecao`. 

## Histogramas

```python
def draw_histplot(num_variables):
    fig, ax = plt.subplots(int(len(num_variables.columns)/2),2, figsize=(15,15))
    colunas = num_variables.columns
    indice = 0
    for j in range(len(ax)):
        for i in range(len(ax[j])):
            ax[j][i].tick_params(labelsize=12)
            ax[j][i] = sns.histplot(data=num_variables[colunas[indice]],ax=ax[j][i])
            indice += 1
    fig.suptitle('Histogramas', position=(.5,1), fontsize=22)
    fig.tight_layout(pad=5)
    return
```

```python
draw_histplot(num_variables_regiao)
```

A avaliação do IDEB ocorre a cada dois anos, através do histograma podemos ratificar isto. Alguns dos insights que extraímos a partir da interpretação dos gráficos: 
* A variável taxa de aprovação possuiu maior frequência de distribuição dos dados para valores próximos a 95;
* O indicador de rendimento apresentou a maior quantidade de dados em valores próximos a 0,95;
* A maioria das notas saeb de matemática e de língua portuguesa estiveram próximas a 250;
* O valor de IDEB mais frequente esteve por volta de 4,8;
* A variável projeção apresentou a maioria dos valores em torno de 4,25.

## KDE plots

O KDEplot é uma forma mais suavizada de exibir os dados em comparação com o histograma. Através desse tipo de gráfico também podemos analisar a distribuição das variáveis.

```python
# Função que auxilia na plotagem dos gráficos KDE
def plot_kde(num_attributes):
    num_attributes.ano = num_attributes.ano.astype('int')
    fig, ax = plt.subplots(int(len(num_attributes.columns)),1, figsize=(12,30))
    colunas = num_attributes.columns
    indice = 0
    for j in range(len(ax)):
      ax[j].tick_params(labelsize=10)
      ax[j] = sns.kdeplot(data=num_attributes[colunas[indice]], ax=ax[j],shade=True)
      indice += 1
    fig.suptitle('KDE Plots', position=(.5,1), fontsize=20)
    fig.tight_layout(pad=5)
    return
# KDE plot 
plot_kde(num_variables_regiao)
```

As interpretações provenientes da análise dos KDEplots foram bem semelhantes às obtidas através dos histogramas.

## Boxplot

O boxplot, ou gráfico de caixa, é um gráfico que permite a identificação de outliers bem como visualizar os quartis. 

```python
# Função que auxilia na plotagem dos gráficos Boxplot
def plot_boxplot(num_attributes):
    num_attributes.ano = num_attributes.ano.astype('int')
    fig, ax = plt.subplots(int(len(num_attributes.columns)/2),2, figsize=(15,20))
    colunas = num_attributes.columns
    indice = 0
    for j in range(len(ax)):
        for i in range(len(ax[j])):
            ax[j][i].tick_params(labelsize=12)
            subtitulo = 'Variável: ' + colunas[indice]
            ax[j][i].tick_params(labelsize=10)
            ax[j][i] = sns.boxplot(data=num_attributes[colunas[indice]], ax=ax[j][i]).set_title(subtitulo)
            indice += 1
    fig.suptitle('Boxplots', position=(.5,1), fontsize=20)
    fig.tight_layout(pad=5)
    return
# Boxplot
plot_boxplot(num_variables_regiao)
```

Notamos que apenas a variável nota saeb matemática possui outlier.
